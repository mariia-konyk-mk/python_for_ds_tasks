```python
A = np.array(
    [[0.18, 0.53],
     [0.75, 0.47],
     [0.35, 0.21]])

B = np.array(
    [[0.17, 0.58, 0.75],
     [0.38, 0.11, 0.15]])

C = np.array([[0], [1]])

D = np.array([[0,1], [1,0]])

display(A, B, C, D)
```

```python
result_matrix= np.dot(B,A)
result_matrix,result_matrix.shape
```

```python
result_matrix_C = np.dot(result_matrix,C)
result_matrix_C
```

```python
result_matrix_D = np.dot(result_matrix,D)
result_matrix_D
```
