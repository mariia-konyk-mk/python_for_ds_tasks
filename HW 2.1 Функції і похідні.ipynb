### Домашні завдання на відпрацювання поняття функції та її похідної

1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp
```

```python
x = sp.symbols('x')
f = 10 * x - 5
derivative_f = sp.diff(f, x)
print("Похідна функції f = 10 * x - 5 = ", derivative_f)
```

```python
x = np.linspace(0, 10, 100)
f = 10 * x - 5
derivative_f =  np.full_like(x,10)
plt.figure(figsize=(10, 5))
plt.plot(x, f, label='f(x) = 10x - 5', color='blue')
plt.plot(x, derivative_f, label="f'(x) = 10", color='green')

plt.title("Графік функції та її похідної")
plt.xlabel("Кіл-ть годин навчання")
plt.ylabel("Оцінка")
plt.legend()
plt.grid(True)

plt.show()
```

Так як це лінійна функція то оцінка постійно зростає лінійно відповідно до кількості годин витрачених на навчання. Похідна становить 10 і показує швидкість з якою одна змінна змінится відповідно до зміни іншої. Так як швидкість не змінюється то швидкість зміни оцінки зі збільшенням часу навчання не збільшується.  

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
x = sp.symbols('x')
f_1 = sp.sin(x)
f_2 = sp.log(x)
derivative_f_1 = sp.diff(f_1, x)
derivative_f_2 = sp.diff(f_2, x)
print("Похідна функції f = sin(x) = ", derivative_f_1)
print("Похідна функції f = ln(x) = ", derivative_f_2)
```

```python
x = np.linspace(0,10,20)
f_1 = np.sin(x)
derivative_f_1 = np.cos(x)


plt.figure(figsize=(10, 5))
plt.plot(x, f_1, label='sin(x)')
plt.plot(x, derivative_f_1, label='cos(x) (похідна)')

plt.legend()
plt.grid(True)
plt.show()
```

```python
x = np.linspace(0,10,20)
f_2 = np.log(x)
derivative_f_2 =  1/x


plt.figure(figsize=(10, 5))
plt.plot(x, f_2, label='log(x))')
plt.plot(x, derivative_f_2, label='1/x (похідна)')

plt.legend()
plt.grid(True)
plt.show()
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
x = sp.symbols('x')
f = x**2 - 5*x +1
derivative_f = sp.diff(f, x)
print("Похідна функції f = 10 * x - 5 = ", derivative_f)
```

```python
x = np.linspace(0, 100, 100)
f = x**2 - 5*x +1
derivative_f = 2*x - 5
plt.figure(figsize=(10, 5))
plt.plot(x, f, label='f(x) = x**2 - 5*x +1', color='blue')
plt.plot(x, derivative_f, label="f'(x) =  2*x - 5", color='green')

plt.title("Графік функції та її похідної")
plt.xlabel("Кількість вироблених одиниць")
plt.ylabel("Загальні витрати")
plt.legend()
plt.grid(True)

plt.show()
```
