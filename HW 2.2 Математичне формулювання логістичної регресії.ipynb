–í —Ü—å–æ–º—É –¥–æ–º–∞—à–Ω—å–æ–º—É –∑–∞–≤–¥–∞–Ω–Ω—ñ –º–∏ —Ä–µ–∞–ª—ñ–∑—É—î–º–æ –ª–æ–≥—ñ—Å—Ç–∏—á–Ω—É —Ä–µ–≥—Ä–µ—Å—ñ—é –Ω–∞ `numpy`.
–¶—ñ –∑–∞–≤–¥–∞–Ω–Ω—è –¥–æ–ø–æ–º–æ–∂—É—Ç—å –≤–∞–º “ë—Ä—É–Ω—Ç–æ–≤–Ω–æ –∑–∞—Å–≤–æ—ó—Ç–∏ –æ—Å–Ω–æ–≤–Ω—ñ –∫–æ–Ω—Ü–µ–ø—Ü—ñ—ó –ª–æ–≥—ñ—Å—Ç–∏—á–Ω–æ—ó —Ä–µ–≥—Ä–µ—Å—ñ—ó —Ç–∞ —Ä–µ–∞–ª—ñ–∑—É–≤–∞—Ç–∏ —ó—Ö –Ω–∞ –ø—Ä–∞–∫—Ç–∏—Ü—ñ üî•

#### –ó–∞–≤–¥–∞–Ω–Ω—è 1: –†–µ–∞–ª—ñ–∑–∞—Ü—ñ—è —Ñ—É–Ω–∫—Ü—ñ—ó —Å–∏–≥–º–æ—ó–¥–∏
1. –ó –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è–º `numpy` –Ω–∞–ø–∏—à—ñ—Ç—å —Ñ—É–Ω–∫—Ü—ñ—é `sigmoid(z)` –¥–ª—è –æ–±—á–∏—Å–ª–µ–Ω–Ω—è –∑–Ω–∞—á–µ–Ω–Ω—è —Å–∏–≥–º–æ—ó–¥–∏ –∑–≥—ñ–¥–Ω–æ –∑ —Ñ–æ—Ä–º—É–ª–æ—é:
   $$
   \sigma(z) = \frac{1}{1 + e^{-z}}
   $$
2. –í–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—é—á–∏ —Ü—é —Ñ—É–Ω–∫—Ü—ñ—é, –æ–±—á–∏—Å–ª—ñ—Ç—å –∑–Ω–∞—á–µ–Ω–Ω—è —Å–∏–≥–º–æ—ó–¥–∏ –¥–ª—è –Ω–∞—Å—Ç—É–ø–Ω–∏—Ö –¥–∞–Ω–∏—Ö: $ z = [-2, -1, 0, 1, 2] $. –í–∏–≤–µ–¥—ñ—Ç—å —Ä–µ–∑—É–ª—å—Ç–∞—Ç –æ–±—á–∏—Å–ª–µ–Ω—å.


```python
import numpy as np
```

```python
def sigmoid(z):
   return 1/ (1+np.e**-z)
```

```python
z = np.array([-2,-1,0,1,2])
```

```python
sigmoid(z)
print(sigmoid(z))
```



#### –ó–∞–≤–¥–∞–Ω–Ω—è 2: –†–µ–∞–ª—ñ–∑–∞—Ü—ñ—è —Ñ—É–Ω–∫—Ü—ñ—ó –≥—ñ–ø–æ—Ç–µ–∑–∏ –¥–ª—è –ª–æ–≥—ñ—Å—Ç–∏—á–Ω–æ—ó —Ä–µ–≥—Ä–µ—Å—ñ—ó
1. –ù–∞–ø–∏—à—ñ—Ç—å —Ñ—É–Ω–∫—Ü—ñ—é `hypothesis(theta, X)`, —è–∫–∞ –æ–±—á–∏—Å–ª—é—î –≥—ñ–ø–æ—Ç–µ–∑—É –¥–ª—è –ª–æ–≥—ñ—Å—Ç–∏—á–Ω–æ—ó —Ä–µ–≥—Ä–µ—Å—ñ—ó, –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—é—á–∏ —Ñ—É–Ω–∫—Ü—ñ—é —Å–∏–≥–º–æ—ó–¥–∏. –§–æ—Ä–º—É–ª–∞ –≥—ñ–ø–æ—Ç–µ–∑–∏:
   $$
   h_\theta(x) = \sigma(\theta^T x) = \frac{1}{1 + e^{-\theta^T x}}
   $$
2. –í–∏–∫–æ—Ä–∏—Å—Ç–∞–π—Ç–µ —Ñ—É–Ω–∫—Ü—ñ—é `hypothesis` –¥–ª—è –æ–±—á–∏—Å–ª–µ–Ω–Ω—è –∑–Ω–∞—á–µ–Ω—å –≥—ñ–ø–æ—Ç–µ–∑–∏ –¥–ª—è –Ω–∞—Å—Ç—É–ø–Ω–∏—Ö –¥–∞–Ω–∏—Ö:
   
   $\theta = [0.5, -0.5]$
   
   $X = \begin{bmatrix} 1 & 2 \\ 1 & -1 \\ 1 & 0 \\ 1 & 1 \end{bmatrix}$

  –í–∏–≤–µ–¥—ñ—Ç—å —Ä–µ–∑—É–ª—å—Ç–∞—Ç –æ–±—á–∏—Å–ª–µ–Ω—å.


```python
theta = np.array([[0.5,-0.5]])
X = np.array([[1, 2], [1, -1], [1, 0], [1,1]])
```

```python
theta.T
```

```python
def hypothesis(theta, X):
    z = np.dot(X, theta.T)
    return sigmoid(z)
```

```python
hypothesis(theta, X)
print(hypothesis(theta, X))
```

#### –ó–∞–≤–¥–∞–Ω–Ω—è 3: –†–µ–∞–ª—ñ–∑–∞—Ü—ñ—è —Ñ—É–Ω–∫—Ü—ñ—ó –¥–ª—è –ø—ñ–¥—Ä–∞—Ö—É–Ω–∫—É –≥—Ä–∞–¥—ñ—î–Ω—Ç—ñ–≤ —Ñ—É–∫–Ω—Ü—ñ—ó –≤—Ç—Ä–∞—Ç
1. –ù–∞–ø–∏—à—ñ—Ç—å —Ñ—É–Ω–∫—Ü—ñ—é `compute_gradient(theta, X, y)`, —è–∫–∞ –æ–±—á–∏—Å–ª—é—î –≥—Ä–∞–¥—ñ—î–Ω—Ç–∏ —Ñ—É–Ω–∫—Ü—ñ—ó –≤—Ç—Ä–∞—Ç –¥–ª—è –ª–æ–≥—ñ—Å—Ç–∏—á–Ω–æ—ó —Ä–µ–≥—Ä–µ—Å—ñ—ó. –§–æ—Ä–º—É–ª–∞ –¥–ª—è –æ–±—á–∏—Å–ª–µ–Ω–Ω—è –≥—Ä–∞–¥—ñ—î–Ω—Ç–∞:
   $$
   \frac{\partial L(\theta)}{\partial \theta_j} = \frac{1}{m} \sum_{i=1}^{m} \left[ (h_\theta(x^{(i)}) - y^{(i)}) x_j^{(i)} \right]
   $$
2. –í–∏–∫–æ—Ä–∏—Å—Ç–∞–π—Ç–µ —Ñ—É–Ω–∫—Ü—ñ—é `compute_gradient` –¥–ª—è –æ–±—á–∏—Å–ª–µ–Ω–Ω—è –≥—Ä–∞–¥—ñ—î–Ω—Ç—ñ–≤ –¥–ª—è –Ω–∞—Å—Ç—É–ø–Ω–∏—Ö –¥–∞–Ω–∏—Ö:

  $\theta = [0.5, -0.5]$

  $X = \begin{bmatrix} 1 & 2 \\ 1 & -1 \\ 1 & 0 \\ 1 & 1 \end{bmatrix}$

  $y = [1, 0, 1, 0]$

  –í–∏–≤–µ–¥—ñ—Ç—å —Ä–µ–∑—É–ª—å—Ç–∞—Ç –æ–±—á–∏—Å–ª–µ–Ω—å.

```python
y=np.array([1,0,1,0])
```

```python
def compute_gradient(theta, X, y):
  m = len(y)
  h = hypothesis(theta, X)
  error = h - y
  return (1/m) * np.dot(X.T, error)
```

```python
compute_gradient(theta, X, y)
```


#### –ó–∞–≤–¥–∞–Ω–Ω—è 4: –†–µ–∞–ª—ñ–∑–∞—Ü—ñ—è –ø–æ–≤–Ω–æ–≥–æ –±–∞—Ç—á –≥—Ä–∞–¥—ñ—î–Ω—Ç–Ω–æ–≥–æ —Å–ø—É—Å–∫—É

**–ó–∞–¥–∞—á–∞:**
1. –ù–∞–ø–∏—à—ñ—Ç—å —Ñ—É–Ω–∫—Ü—ñ—é `full_batch_gradient_descent(X, y, lr=0.1, epochs=100)`, —è–∫–∞ —Ä–µ–∞–ª—ñ–∑—É—î –∞–ª–≥–æ—Ä–∏—Ç–º Full –≥—Ä–∞–¥—ñ—î–Ω—Ç–Ω–æ–≥–æ —Å–ø—É—Å–∫—É –¥–ª—è –ª–æ–≥—ñ—Å—Ç–∏—á–Ω–æ—ó —Ä–µ–≥—Ä–µ—Å—ñ—ó. –í–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É–π—Ç–µ —Ç–∞–∫—ñ —Ñ–æ—Ä–º—É–ª–∏:
   - –ì—ñ–ø–æ—Ç–µ–∑–∞: $ h_\theta(x) = \sigma(\theta^T x) $
   - –û–Ω–æ–≤–ª–µ–Ω–Ω—è –ø–∞—Ä–∞–º–µ—Ç—Ä—ñ–≤: $ \theta_j := \theta_j - \alpha \frac{\partial L(\theta)}{\partial \theta_j} $
2. –í–∏–∫–æ—Ä–∏—Å—Ç–∞–π—Ç–µ —Ñ—É–Ω–∫—Ü—ñ—é `full_batch_gradient_descent` –¥–ª—è –æ–±—á–∏—Å–ª–µ–Ω–Ω—è –ø–∞—Ä–∞–º–µ—Ç—Ä—ñ–≤ –º–æ–¥–µ–ª—ñ –Ω–∞ –Ω–∞—Å—Ç—É–ø–Ω–∏—Ö –¥–∞–Ω–∏—Ö:

  $X = \begin{bmatrix} 1 & 2 \\ 1 & -1 \\ 1 & 0 \\ 1 & 1 \end{bmatrix}$

  $y = [1, 0, 1, 0]$

  –£–≤–∞–≥–∞! –ú–∞—Ç—Ä–∏—Ü—è $X$ –≤–∂–µ –º–∞—î —Å—Ç–æ–≤–ø–µ—Ü—å –æ–¥–∏–Ω–∏—Ü—å —ñ –ø–µ—Ä–µ–¥–±–∞—á–∞—î—Ç—å—Å—è, —â–æ —Ü–µ. - —Å—Ç–æ–≤–ø–µ—Ü—å –¥–ª—è intercept - –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ –∑—Å—É–≤—É.

  –í–∏–≤–µ–¥—ñ—Ç—å —Ä–µ–∑—É–ª—å—Ç–∞—Ç –æ–±—á–∏—Å–ª–µ–Ω—å.


```python
def full_batch_gradient_descent(X, y, lr=0.1, epochs=100):
    theta = np.array([0, 0])

    for epoch in range(epochs):
        gradient = compute_gradient(theta, X, y)
        theta = theta - lr * gradient

    return theta
```

```python
theta_j = full_batch_gradient_descent(X, y, lr=0.1, epochs=100)
print(theta_j )
```

#### –ó–∞–≤–¥–∞–Ω–Ω—è 5. –û–±—á–∏—Å–ª–µ–Ω–Ω—è —Ç–æ—á–Ω–æ—Å—Ç—ñ –º–æ–¥–µ–ª—ñ

1. –ù–∞–ø–∏—à—ñ—Ç—å —Ñ—É–Ω–∫—Ü—ñ—é `predict_proba(theta, X)`, —è–∫–∞ –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—î –∑–Ω–∞–π–¥–µ–Ω—ñ –ø–∞—Ä–∞–º–µ—Ç—Ä–∏ $\theta$ –¥–ª—è –æ–±—á–∏—Å–ª–µ–Ω–Ω—è –π–º–æ–≤—ñ—Ä–Ω–æ—Å—Ç–µ–π –Ω–∞–ª–µ–∂–Ω–æ—Å—Ç—ñ –ø–æ—Ç–æ—á–Ω–æ–≥–æ –ø—Ä–∏–∫–ª–∞–¥—É –∑ –¥–∞–Ω–∏—Ö –¥–æ –∫–ª–∞—Å—É $y=1$ –Ω–∞ –æ—Å–Ω–æ–≤—ñ –∑–Ω–∞—á–µ–Ω—å $\sigma(\theta^T x)$.

2. –ù–∞–ø–∏—à—ñ—Ç—å —Ñ—É–Ω–∫—Ü—ñ—é `predict(theta, X, threshold=0.5)`, —è–∫–∞ –æ–±—á–∏—Å–ª—é—î –∫–ª–∞—Å –∑ –ø–µ—Ä–µ–¥–±–∞—á–µ–Ω–æ—ó —ñ–º–æ–≤—ñ—Ä–Ω–æ—Å—Ç—ñ –Ω–∞–ª–µ–∂–Ω–æ—Å—Ç—ñ –µ–∫–∑–µ–º–ø–ª—è—Ä–∞ –¥–æ –∫–ª–∞—Å—É 1 –∑ –ø–æ—Ä–æ–≥–æ–º 0.5. –¢–æ–±—Ç–æ —è–∫—â–æ –π–º–æ–≤—ñ—Ä–Ω—ñ—Å—Ç—å –º–µ–Ω—à–µ 0.5, —Ç–æ –ø–µ—Ä–µ–¥–±–∞—á–∞—î–º–æ –∫–ª–∞—Å 0, —ñ–Ω–∞–∫—à–µ –∫–ª–∞—Å 1.

3. –ù–∞–ø–∏—à—ñ—Ç—å —Ñ—É–Ω–∫—Ü—ñ—é `accuracy(y_true, y_pred)`, —è–∫–∞ –æ–±—á–∏—Å–ª—é—î —Ç–æ—á–Ω—ñ—Å—Ç—å –º–æ–¥–µ–ª—ñ, –≤–∏–∑–Ω–∞—á–∏–≤—à–∏ —á–∞—Å—Ç–∫—É –ø—Ä–∞–≤–∏–ª—å–Ω–æ –ø–µ—Ä–µ–¥–±–∞—á–µ–Ω–∏—Ö –∫–ª–∞—Å—ñ–≤.

  –§–æ—Ä–º—É–ª–∞ –º–µ—Ç—Ä–∏–∫–∏ Accuracy:
  $$
  \text{Accuracy} = \frac{\sum_{i=1}^{m} I(\hat{{y}^{(i)}} = y^{(i)})}{m}
  $$

  –¥–µ $\hat{{y}^{(i)}}$ - –ø–µ—Ä–µ–¥–±–∞—á–µ–Ω–µ –∑–Ω–∞—á–µ–Ω–Ω—è –∫–ª–∞—Å—É, $I$ - —ñ–Ω–¥–∏–∫–∞—Ç–æ—Ä–Ω–∞ —Ñ—É–Ω–∫—Ü—ñ—è (—è–∫–∞ –¥–æ—Ä—ñ–≤–Ω—é—î 1, —è–∫—â–æ —É–º–æ–≤–∞ –≤–∏–∫–æ–Ω—É—î—Ç—å—Å—è, —ñ 0 - —è–∫—â–æ –Ω—ñ), $m$ - –∫—ñ–ª—å–∫—ñ—Å—Ç—å –ø—Ä–∏–∫–ª–∞–¥—ñ–≤.

4. –û–±—á–∏—Å–ª—ñ—Ç—å –∑ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è–º –¥–∞–Ω–∏—Ö –≤ –∑–∞–≤–¥–∞–Ω–Ω—ñ 4 $X$, $y$ —Ç–∞ –æ–±—á–∏—Å–ª–µ–Ω–∏—Ö –∫–æ–µ—Ñ—ñ—Ü—ñ—î–Ω—Ç–∞—Ö $\theta$ —Ç–∞ –≤–∏–≤–µ–¥—ñ—Ç—å –Ω–∞ –µ–∫—Ä–∞–Ω:
  - –ø–µ—Ä–µ–¥–±–∞—á–µ–Ω—ñ –º–æ–¥–µ–ª–ª—é —ñ–º–æ–≤—ñ—Ä–Ω–æ—Å—Ç—ñ –Ω–∞–ª–µ–∂–Ω–æ—Å—Ç—ñ –∫–æ–∂–Ω–æ–≥–æ –∑ –µ–∫–∑–µ–º–ø–ª—è—Ä—ñ–≤ –≤ –º–∞—Ç—Ä–∏—Ü—ñ `X` –¥–æ –∫–ª–∞—Å—É 1
  - –∫–ª–∞—Å–∏ –∫–æ–∂–Ω–æ–≥–æ –µ–∫–∑–µ–º–ø–ª—è—Ä–∞ –∑ –º–∞—Ç—Ä–∏—Ü—ñ `X`
  - —Ç–æ—á–Ω—ñ—Å—Ç—å –º–æ–¥–µ–ª—ñ.

```python
def predict_proba(theta, X):
    z = np.dot(X, theta.T)
    return sigmoid(z)
```

```python
predict_proba = predict_proba (theta_j, X)
predict_proba
```

```python
def predict(theta, X, threshold=0.5):
    return (predict_proba >= threshold).astype(int)
```

```python
predict= predict(theta_j, X, threshold=0.5)
print(predict)
```

```python
def accuracy(y_true, y_pred):
    m = len(y)
    return np.sum(y_true == y_pred)/  m
```

```python
accuracy= accuracy(y, predict)
accuracy
```
